```python
from imblearn.over_sampling import SMOTE

# Resampling the dataset using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
# Verify the class distribution after SMOTE
sns.countplot(x=y_resampled)
plt.title("Class Distribution After SMOTE")
plt.show()
```

```python
data['Time'] = pd.to_datetime(data['Time'])
```

```python
def time_based_features(data):
    data['transaction_time'] = pd.to_datetime(data['Time'], unit='s')
    data['hour'] = data['transaction_time'].dt.hour
    data['day_of_week'] = data['transaction_time'].dt.dayofweek
    return data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
# Splitting the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# Define the models
models = {
    'Random Forest': RandomForestClassifier(),
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(),
    'XGB': XGBClassifier()
}
```

```python
for model_name, model in models.items():
    model.fit(X_train, y_train)
    print(f"Trained {model_name} on training data.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

for model_name, model in models.items():
    y_pred = model.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    print(f"Model: {model_name}")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("\n")
```
